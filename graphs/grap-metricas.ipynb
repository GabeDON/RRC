```python
### import matplotlib.pyplot as plt

# Dados
meses = ['Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
downloads = [14, 8, 12, 7, 9, 10, 21]

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(meses, downloads, color='skyblue')

# Adicionar título e rótulos
plt.title('Downloads Mensais')
plt.xlabel('Mês')
plt.ylabel('Número de Downloads')

# Mostrar o gráfico
plt.savefig('dowloads-mensais.png',format='png')
plt.show()
```
